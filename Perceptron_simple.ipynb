Un perceptrón es un algoritmo de aprendizaje supervisado para la clasificación binaria. Es el bloque de construcción más simple en el campo de las redes neuronales. Matemáticamente, un perceptrón toma un vector de entrada $ \textbf{x} $, lo multiplica por un conjunto de pesos $ \textbf{w} $, suma un sesgo $ b $, y luego aplica una función de activación. La operación se puede resumir como:

$
f(\textbf{x}) = \text{Activación}(\textbf{w} \cdot \textbf{x} + b)
$

Donde $ \textbf{w} \cdot \textbf{x} $ es el producto punto entre los vectores de entrada y peso, y $ \text{Activación} $ es una función de activación como la función escalón, sigmoide, etc.

El objetivo del entrenamiento de un perceptrón es encontrar los valores de $ \textbf{w} $ y $ b $ que permiten clasificar correctamente los datos de entrada. Esto se hace mediante la iteración a través de un conjunto de entrenamiento y la actualización de los pesos y sesgos de acuerdo con alguna regla de aprendizaje, como la regla de aprendizaje del perceptrón.

Cabe destacar que un perceptrón simple solo es capaz de resolver problemas que son linealmente separables. Para problemas más complejos, se suelen utilizar redes neuronales multicapa, que consisten en múltiples perceptrones organizados en capas.

## *Visualización gráfica de la compuerta AND*

```python
import matplotlib.pyplot as plt

# Puntos con etiqueta 0
x0 = [0, 0, 1]
y0 = [0, 1, 0]

# Puntos con etiqueta 1
x1 = [1]
y1 = [1]

# Crear el gráfico
plt.figure(figsize=(6, 6))

# Dibujar puntos
plt.scatter(x0, y0, c='b', marker='o', label='Etiqueta 0')
plt.scatter(x1, y1, c='r', marker='x', label='Etiqueta 1')

# Configurar gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Compuerta AND')
plt.grid(True)
plt.legend()

# Mostrar gráfico
plt.show()
```

## Perceptron compuerta AND

```python
import numpy as np

# Datos de entrada y etiquetas
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# Inicialización de pesos y sesgo
weights = np.zeros(2)
bias = 5
learning_rate = 0.3
epochs = 60

# Función de activación (función escalón)
def activation_function(x):
    return 1 if x >= 0 else 0

# Entrenamiento del perceptrón
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = activation_function(linear_output)

        # Actualización de pesos y sesgo
        error = y[i] - prediction
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

# Mostrar los resultados
print("Pesos finales:", weights)
print("Sesgo final:", bias)

# Prueba del modelo entrenado
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    prediction = activation_function(linear_output)
    print(f"Para la entrada {X[i]}, la predicción es {prediction}")
```

## *Datos linealmente separables*

Los datos linealmente separables son aquellos que se pueden separar completamente en clases distintas utilizando una sola línea recta (en el caso de datos en dos dimensiones), un plano (en tres dimensiones), o un hiperplano (en más de tres dimensiones). En términos más formales, un conjunto de datos $ \{(x_1, y_1), (x_2, y_2), \ldots, (x_n, y_n)\} $ es linealmente separable si existe un vector de pesos $ \textbf{w} $ y un sesgo $ b $ tales que:

$
y_i = \begin{cases}
1 & \text{si } \textbf{w} \cdot \textbf{x}_i + b > 0 \\
0 & \text{si } \textbf{w} \cdot \textbf{x}_i + b \leq 0
\end{cases}
$

para cada $ i = 1, 2, \ldots, n $.

En otras palabras, si es posible trazar una línea, plano o hiperplano que separe todas las muestras de una clase de todas las muestras de otra sin errores, entonces los datos son linealmente separables.



```python
import matplotlib.pyplot as plt

# Puntos con etiqueta 0
x0 = [0, 1]
y0 = [0, 1]

# Puntos con etiqueta 1
x1 = [0, 1]
y1 = [1, 0]

# Crear el gráfico
plt.figure(figsize=(6, 6))

# Dibujar puntos
plt.scatter(x0, y0, c='b', marker='o', label='Etiqueta 0')
plt.scatter(x1, y1, c='r', marker='x', label='Etiqueta 1')

# Configurar gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Compuerta XOR')
plt.grid(True)
plt.legend()

# Mostrar gráfico
plt.show()
```

Supongamos que los datos son linealmente separables. Entonces, debería existir un hiperplano que los separe perfectamente. En el caso de dos dimensiones, este "hiperplano" se reduce a una línea recta. La ecuación de esta línea podría escribirse como:

$
w_1 x_1 + w_2 x_2 + b = 0
$

Donde $ w_1 $ y $ w_2 $ son los pesos y $ b $ es el sesgo.

Ahora, si los datos son linealmente separables, entonces los puntos de cada clase deberían caer de un solo lado de este hiperplano. Es decir, si tomamos dos puntos $ \textbf{a} $ y $ \textbf{b} $ de la misma clase, cualquier punto en la línea que conecta $ \textbf{a} $ y $ \textbf{b} $ también debería pertenecer a esa clase. Matemáticamente, esto se describe como:

$
\textbf{z} = \lambda \textbf{a} + (1 - \lambda) \textbf{b}
$

Donde $ \lambda $ es un número real entre 0 y 1.

Para la compuerta XOR, consideremos los puntos (0,0) y (1,1), que ambos tienen la etiqueta 0. De acuerdo con nuestra suposición inicial de que los datos son linealmente separables, cualquier punto en la línea entre estos dos puntos también debería tener la etiqueta 0.

Si escogemos $ \lambda = 0.5 $, obtenemos:

$
\textbf{z} = 0.5 \times (0,0) + 0.5 \times (1,1) = (0.5, 0.5)
$

El punto (0.5, 0.5) debería tener, por lo tanto, la etiqueta 0. Sin embargo, según la definición de la compuerta XOR, este punto tendría la etiqueta 1. Esto es una contradicción, lo que implica que nuestra suposición inicial de que los datos son linealmente separables debe ser incorrecta.

Por lo tanto, los datos de la compuerta XOR no son linealmente separables.

```python
import numpy as np

# Datos de entrada y etiquetas
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

# Inicialización de pesos y sesgo
weights = np.random.rand(2)
bias = np.random.rand(1)
learning_rate = 0.1
epochs = 10000

# Función de activación (función escalón)
def activation_function(x):
    return 1 if x >= 0 else 0

# Entrenamiento del perceptrón
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = activation_function(linear_output)

        # Actualización de pesos y sesgo
        error = y[i] - prediction
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

# Mostrar los resultados
print("Pesos finales:", weights)
print("Sesgo final:", bias)

# Prueba del modelo entrenado
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    prediction = activation_function(linear_output)
    print(f"Para la entrada {X[i]}, la predicción es {prediction}")
```

# Pseudocódigo para aplicar el Perceptrón a un conjunto de datos simulados

## 1. Importar librerías
- Importar Numpy para operaciones matriciales
- Importar Matplotlib para visualización

## 2. Generar datos simulados
- Fijar la semilla del generador de números aleatorios para reproducibilidad
- Generar 100 puntos en 2D usando números aleatorios (`X`)
- Asignar etiquetas (`y`) basadas en si la suma de las coordenadas de un punto es mayor que 10

## 3. Inicializar variables
- Inicializar pesos y sesgo con valores aleatorios (`weights`, `bias`)
- Establecer la tasa de aprendizaje (`learning_rate`)
- Establecer el número de épocas (`epochs`)

## 4. Definir la función de activación
- Usar la función escalón unitario como función de activación

Una función de activación en el contexto de las redes neuronales artificiales y modelos similares, como el perceptrón, es una función matemática aplicada a una suma ponderada de las entradas y un sesgo. El objetivo de la función de activación es transformar la entrada recibida para producir una salida que se utilizará posteriormente. Esta salida puede ser la predicción final en el caso de una capa de salida, o puede ser la entrada para las neuronas en las siguientes capas de una red neuronal más compleja.


Algunas de las funciones de activación más comúnmente utilizadas incluyen:

- **Función escalón unitario**: Es una función binaria que devuelve 0 si la entrada es menor que 0 y 1 en caso contrario. Se utiliza en el perceptrón simple.
  
- **Sigmoidal**: Toma un valor de entrada y lo transforma en una salida en el rango entre 0 y 1. Su fórmula es $ \sigma(x) = \frac{1}{1 + e^{-x}} $.

- **Tangente hiperbólica (tanh)**: Similar al sigmoide, pero varía en el rango de -1 a 1.

- **Rectificadora (ReLU)**: Devuelve 0 si la entrada es negativa y la entrada misma si es positiva. Su fórmula es $ f(x) = \max(0, x) $.

Cada una de estas funciones de activación tiene sus propias ventajas y desventajas, y la elección de una función de activación específica depende del problema que se esté tratando de resolver.

## 5. Entrenar el Perceptrón
- Para cada época hacer:
    - Para cada punto en `X` hacer:
        1. Calcular la suma ponderada de las entradas (`linear_output`)
        2. Aplicar la función de activación para obtener la predicción (`prediction`)
        3. Calcular el error (`error`)
        4. Actualizar los pesos y el sesgo

## 6. Probar el Perceptrón entrenado
- Inicializar una lista vacía para las predicciones (`predictions`)
- Para cada punto en `X` hacer:
    1. Calcular la suma ponderada de las entradas
    2. Aplicar la función de activación para obtener la predicción
    3. Añadir la predicción a la lista de predicciones

## 7. Visualizar resultados
- Usar Matplotlib para dibujar los puntos y la frontera de decisión
- Mostrar el gráfico



```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos simulados
np.random.seed(42)
X = np.random.rand(100, 2) * 10  # 100 puntos en 2D
y = np.array([1 if x[0] + x[1] > 10 else 0 for x in X])  # Etiquetas: 1 si la suma de las coordenadas > 10, 0 en caso contrario

# Inicializar pesos y sesgo
weights = np.random.rand(2)
bias = np.random.rand(1)
learning_rate = 0.1
epochs = 10

# Función de activación
def activation_function(x):
    return 1 if x >= 0 else 0

# Entrenar el perceptrón
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = activation_function(linear_output)
        error = y[i] - prediction
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

# Probar el perceptrón entrenado
predictions = []
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    prediction = activation_function(linear_output)
    predictions.append(prediction)

# Visualizar los datos y la frontera de decisión
plt.figure(figsize=(10, 10))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = np.array([activation_function(np.dot(np.array([x, y]), weights) + bias) for x, y in zip(np.ravel(xx), np.ravel(yy))])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm')
plt.title('Perceptrón en datos simulados')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
